```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


df = pd.read_csv('CCTV정보.csv', encoding='cp949', low_memory=False)
```

```python
#전체 값이 몇개가 있는지 확인하는 방법은? -> 모양을 생각하자.
df.shape
```

```python
#그렇다면 이번에는 데이터셋에 어떤 제목의 컬럼이 있는지를 알아보려면 어떻게 해야할까?
df.columns
```

```python
#만약 특정 열의 정보만 출력하려고 하면 어떻게 해야할까?
df_1 = df[['소재지도로명주소', '관리기관전화번호']]

df_1
```

```python
#그렇다면 이번에는 소재지도로명주소에 있는 NaN값을 지운 후의 상태를 보도록 하자.
df_2 = df_1.dropna(subset='소재지도로명주소')

df_2
```

```python
#그렇다면 이번에는 소재지 기준 '서울특별시' 포함하는 행만 남기는 처리를 해보자.
df_3 = df_2[df_2['소재지도로명주소'].str.contains('서울특별시')]

df_3
```

```python
#시군구 열 추가 loc 소재지전체주소[1] => 구별[1]로 CCTV가 몇개씩 있는가?
df_3.loc[:, '시군구'] = df_3['소재지도로명주소'].str.strip().str[1]

df_3
```

```python
#그룹으로 묶기 -> '시군구', 크기, 카메라대수
df_4 = df_3.groupby('시군구').size().reset_index(name='카메라대수')

df_4
```

```python
#상위 30개만 출력해서 이를 chart에 반영할 수 있도록
df_top_30 = df_4.sort_values('카메라대수', ascending=False).head(30)

df_top_30
```

```python
#바 차트로 만들기
plt.bar(
    df_top_30['카메라대수'],
    df_top_30['시군구'],
    linestyle='-',
    color='b'
)

#제목
plt.title('시군구별 카메라 대수')

#크기 조절
plt.figure(figsize=(10,8))

#x축
plt.xlabel('시군구')

#y축
plt.ylabel('카메라 대수')
```
