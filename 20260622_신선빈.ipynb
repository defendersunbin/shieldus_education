### 연습문제 풀이 1
### 과목별 점수의 총합과 평균 및 총점과 평균 출력

```python
#연습문제 풀이 1
#과목별 점수의 총합과 평균 및 총점과 평균 출력
language = 85
math = 70
english = 95

#국어, 영어, 수학 점수의 총합과 평균 구하기
total_score = sum([language, math, english])
avg_score = total_score // 3
print(f"과목별 점수의 총합: {total_score}")
print(f"과목별 점수의 평균: {avg_score}")

#각 과목의 점수와 총점, 평균 출력하기
print("과목별 점수 ->", f"국어: {language}, 수학: {math}, 영어: {english}")
```

### 연습문제 풀이 2
### 성별이 남성이고 18세 이상의 성인 여부를 판단하는 조건식

```python

gender = 'male'
age = 18

if(gender == 'male' and age >= 18):
    print(True)
elif(gender == 'female' and age >= 18):
    print(False)
else:
    print(False)
```
