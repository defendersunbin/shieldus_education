# 고객 구매 데이터 마케팅 분석 시스템

```python
#필요한 라이브러리 호출
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 한글 폰트 설정
# Windows
# plt.rcParams['font.family'] = 'Malgun Gothic'

# Mac
plt.rcParams['font.family'] = 'AppleGothic'  # Mac 한글 폰트
plt.rcParams['axes.unicode_minus'] = False

#먼저 CustomerSalesAnalysis class 정의
class CustomerSalesAnalysis():
    def __init__ (self):
        #초기 함수
        self.df = pd.DataFrame({
            '고객명': np.random.choice([f'고객{i}' for i in range(1,11)], size=100),
            '구매일자': pd.date_range(start='2024-01-01', periods=100, freq='D'),
            '상품명': np.random.choice(['상품A', '상품B', '상품C'], size=100),
            '수량': np.random.randint(1, 10, size=100),
            '단가': np.random.randint(1000, 50000, size=100)
        })
        self.df['총매출'] = self.df['수량'] * self.df['단가']

    #월별 매출 총합 막대 그래프
    def summarize(self):
        #월 행
        self.df['월'] = self.df['구매일자'].dt.month

        #월별 총매출 합계
        monthly = self.df.groupby('월')['총매출'].sum()

        #막대 그래프
        plt.figure(figsize=(8, 5))
        #바 그래프
        plt.bar(monthly.index, monthly.values, color='skyblue', width=0.4)
        #제목
        plt.title('월별 매출 총합')
        #x축
        plt.xlabel('월')
        #y축
        plt.ylabel('매출')
        #결과로 출력
        plt.show()

    #고객별 누적 매출 pie chart
    def plot_customer_pie(self):
        #고객별 누적 총합계 (sum)
        customer = self.df.groupby('고객명')['총매출'].sum()

        #차트 그리기
        plt.figure(figsize=(8,5))
        #파이 차트
        plt.pie(customer.values, labels=customer.index, autopct='%1.1f%%')
        #차트 제목
        plt.title('고객별 누적 매출')
        #결과 출력
        plt.show()
```

```python
csa = CustomerSalesAnalysis()
csa.summarize()
csa.plot_customer_pie()
```
