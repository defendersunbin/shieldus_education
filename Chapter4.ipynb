# 기초통계

```python
#pandas를 이용 => 평균, 중앙, 최빈값
#mean, median, mode
import pandas as pd
```

```python
data = [10, 15, 10, 20, 30, 10, 40]

data_series = pd.Series(data)
```

```python
#평균
mean_val = data_series.mean()
#중앙값
median_val = data_series.median()
#최빈값(만약 그 중 첫번째 값을 알고 싶다면 mode()[0]을 붙이면 된다)
mode_val = data_series.mode()[0]
```

```python
print(mean_val)
print(median_val)
print(mode_val)
```

# 데이터 분산성(분산, 표준편차, 사분위수)

```python
data = {'점수': [55, 60, 65, 70, 75, 80, 85, 90, 95]}

df = pd.DataFrame(data)
```

```python
#분산 - var
var_value = df['점수'].var()
print(var_value)
```

```python
#표준편차 - std
std_value = df['점수'].std()
print(std_value)
```

```python
#사분위수
q1_value_pd = df['점수'].quantile(0.25) #1사분위수
q2_value_pd = df['점수'].quantile(0.5) #2사분위수
q3_value_pd = df['점수'].quantile(0.75) #3사분위수
q4_value_pd = df['점수'].max() #최대값

print(q1_value_pd)
print(q2_value_pd)
print(q3_value_pd)
print(q4_value_pd)

#IQR (Q3-Q1)
iqr_value = q3_value_pd - q1_value_pd
print(iqr_value)
```

# 서울에 있는 스타벅스 매장 위치 분석

```python
df_1 = df[['번호', '영업상태명', '사업장명', '소재지전체주소', '도로명전체주소']]
df_1
```

```python
#소재지 기준 값 => NaN의 행을 삭제
df_2 = df_1.dropna(subset=['소재지전체주소'])
df_2
```

```python
#소재지 기준 => '서울특별시' 포함하는 행만 남기는 처리
df_3 = df_2[df_2['소재지전체주소'].str.contains('서울특별시')]
df_3
```

```python
df_3.info()
```

```python
df_4 = df_3[df_3['사업장명'].str.contains('스타벅스')]
df_4
```

```python
df_5 = df_4[df_4['영업상태명'] != '폐업']
df_5
```

```python
#시군구 열 추가 소재지전체주소[1] => 구별로 매장이 몇개씩 있는지?
df_5.loc[:, '시군구'] = df_5['소재지전체주소'].str.split().str[1]
df_5
```

```python
#~별 -> 그룹핑
df_6 = df_5.groupby('시군구').size().reset_index(name='사업장수')
df_6
```

```python
plt.bar(
    df_6['시군구'],
    df_6['사업장수']
)
plt.xticks(rotation=45)
plt.title('서울특별시 스타벅스 사업장 분석표')
plt.xlabel('시군구')
plt.ylabel('사업장 수')
```
