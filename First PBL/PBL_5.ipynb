# 학생 과목별 성적 통계 분석 시스템

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

# 한글 폰트 설정
# Windows
# plt.rcParams['font.family'] = 'Malgun Gothic'

# Mac
plt.rcParams['font.family'] = 'AppleGothic'  # Mac 한글 폰트
plt.rcParams['axes.unicode_minus'] = False

#1. StudentScoreAnalysis 클래스 정의
class StudentScoreAnalysis():
    #초기 함수 정의(이름, 과목, 점수)
    def __init__(self):
        self.df = pd.DataFrame({
            "이름": [f'학생{i}' for i in range(1, 21)],
            "수학": np.random.randint(low=50, high=100, size=20),
            "영어": np.random.randint(low=50, high=100, size=20),
            "과학": np.random.randint(low=50, high=100, size=20),
        })
    
    #학생 과목별 성적 - 각 50~100 사이 난수
    def plot_subject_average(self):
        #과목별 평균 계산
        avg = self.df[['수학', '영어', '과학']].mean()

        #막대 그래프 - bar()
        plt.figure(figsize=(8,5))
        #x축 변수, y축 변수, 표시 색깔, 과목별 다른 색
        colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink']
        random.shuffle(colors)  # 실행마다 순서 섞기
        plt.bar(avg.index, avg.values, color=colors[:3], width=0.4)
        #plot 제목
        plt.title('과목별 평균 점수')
        #x축
        plt.xlabel('과목')
        #y축
        plt.ylabel('평균 점수')
        #출력
        plt.show()

    
    #평균 계산 및 정렬: mean() + sort_values()
    def plot_top5_students(self):
        #학생별 평균 계산
        self.df['평균'] = self.df[['수학', '영어', '과학']].mean(axis=1)

        #상위 5명 추출
        top5 = self.df.sort_values('평균', ascending=False).head(5)

        #막대 그래프
        plt.figure(figsize=(8,5))
        #x축 객체, y축 객체, 표시 색깔
        plt.bar(top5['이름'], top5['평균'],
        color=['red', 'blue', 'green', 'orange', 'purple'],  # 학생별 다른 색
        width=0.4)  # 막대 두께 조절
        #표 제목
        plt.title('평균 성적 상위 5명')
        #x축
        plt.xlabel('학생')
        #y축
        plt.ylabel('평균 점수')
        #결과 출력
        plt.show()
```

```python
sea = StudentScoreAnalysis()
sea.plot_subject_average()
sea.plot_top5_students()
```
