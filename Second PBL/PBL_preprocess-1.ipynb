# PBL-1: 당뇨병 진단 데이터를 활용한 데이터 전처리 및 EDA 분석

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder

#데이터 불러오기
df = pd.read_csv('diabetes.csv')
```

```python
#결측치 처리하기
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[cols_with_zero] = df[cols_with_zero].replace(0, np.nan)

for cols in cols_with_zero:
    df[cols] = df[cols].fillna(df[cols].mean())

print(df[cols_with_zero].isnull().sum())
```

```python
#이상치 처리하기
for cols in ['SkinThickness', 'Insulin']:
    overreact = df[cols].quantile(0.99)
    df.loc[df[cols] > overreact, cols] = df[cols].mean()

print(df[['SkinThickness', 'Insulin']].describe())
```

```python
#Age열을 MinMax로 정규화
minmax = MinMaxScaler()
df[['Age']] = minmax.fit_transform(df[['Age']])

print(df['Age'].min(), df['Age'].max())
```

```python
#EDA:
#1. 각 열의 결측치 개수 출력
print(df.isnull().sum())
```

```python
#2. Outcome 별 Glucose 평균 출력
print(df.groupby('Outcome')['Glucose'].mean())
```
