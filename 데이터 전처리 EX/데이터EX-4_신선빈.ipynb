# 데이터 전처리 EX-4

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder

data = pd.read_csv('adult.csv')

data
```

```python
#Label encoding
label_encoder = LabelEncoder()
categorical_cols = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']
for col in categorical_cols:
    data[col + '_encoded'] = label_encoder.fit_transform(data[col])

#onehot encoding
label_onehot_encoder = OneHotEncoder(sparse_output=False)
onehot_result = label_onehot_encoder.fit_transform(data[categorical_cols])

#결과를 dataframe으로 변환
one_hot_columns = label_onehot_encoder.get_feature_names_out()
one_hot_df = pd.DataFrame(onehot_result, columns=one_hot_columns)

#기존 데이터셋에 추가
data = pd.concat([data, one_hot_df], axis=1)
print(data)
```
