# Loop

### 1부터 100까지의 합을 구하는 프로그램 작성

```python
#for문
sum = 0
for i in range(1, 101):
    sum += i
print(f'총합: {sum}')
```

```python
#while문
num = 0
sum = 0
while num < 100:
    num += 1
    sum += num

print(f'총합: {sum}')
```

### for문을 이용해 1부터 10까지를 곱한 결과를 출력하는 프로그램 작성

```python
#for문
sum = 1
for i in range(1, 11):
    sum *= i
print(f'총합: {sum}')
```

### 구구단의 짝수단(2,4,6,8)만 출력하는 프로그램 작성

```python
for dan in range(2, 9, 2):
    print(dan, "단")
    for i in range(1, dan+1):
        print(f'{dan} X {i} = {dan * i}')
```

### while문을 무한 루프로 구성해 작성하는 예제

```python
#1부터 시작해서 모든 짝수와 3의 배수를 더해서 그 합이 언제 1000이 넘어서는지, 그리고 1000이 넘어선 값은 얼마가 되는지 계산해서 출력
num = 0
sum = 0

while True:
    num += 1
    #1. 모든 짝수와 3의 배수를 더한다.
    if num%2==0 or num%3==0:
        sum += num
    
    #2. 합이 언제 1000을 넘어서는지, 1000을 넘어선 값은 얼마가 되는지 그리고 break문으로 끝내 결과 도출
    if sum > 1000:
        print(f'합이 1000을 넘어서는 숫자: {num}')
        print(f'1000을 넘어선 값은 얼마가 되는지: {sum}')
        break
```
