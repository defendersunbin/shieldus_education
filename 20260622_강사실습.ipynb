```python
# 파이썬의 데이터 (숫자, 문자열, 논리값) => 객체 형식으로 저장!
# 변수가 가르키는것을 메모리 주소값
# 객체 : 변수, 함수들
# 변수.함수()
num = 10 # int 정수
num1 = 3.14 # float 실수 
str_10 = '10'  # str 문자열
```

```python
# 문자열생성 =>  ' ' " "  ''' ''' """ """
str_1 = 'hello~'
str_2 = "안녕하세요"
str_3 = '''
첫번째 라인
2
3
네번째 라인
'''
str_4 = """
select *
from emp
where deptno=10
"""
```

```python
print(str_1)
print(str_2)
print(str_3)
print(str_4, type(str_4))
```

```python
# 문자열의 기본 조작
# 덧샘, 곱샘, 문자의 개수
print('hello~'+' python')
print('='*20)
```

```python
len(str_4)  # 시퀀스 데이터 => 문자열, List, Tuple
```

```python
# 문자열의 인덱싱
str_5 = 'hello~ python!!'
# index=>01234567891011121314      
print(len(str_5))
```

```python
# 문자열(시퀀스 데이터)에서 특정위치의 문자 추출
print(str_5[0])
print(str_5[-1])
print(str_5[14])
```

```python
print(str_5)
print(str_5[0:5])
print(str_5[:6])
print(str_5[5:])
```

```python
# 문자열 포맷팅
# 3.6 버전 이후 => f''
user_name = '손흥민'
user_age = 20
```

```python
# 안녕하세요! 저는 000 입니다. 나이는 00살 입니다.
print('안녕하세요! 저는 '+user_name+'입니다. 나이는 '+str(user_age)+'살 입니다.')
print(f'안녕하세요! 저는 {user_name} 입니다. 나이는 {user_age}살 입니다.')
```
