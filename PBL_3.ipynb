# OOP 기반 학생 성적 분석 시스템

```python
#1. StudentScores 클래스 정의
class StudentScores():
    #초기 함수(파일 읽기)
    def __init__(self, filepath):
        self.filepath = filepath
        self.scores = {}
        self.load_scores()

    #파일에서 학생들의 시험 점수 불러오기
    def load_scores(self):
        f = open('scores_korean.txt', 'r')
        #데이터에 있는 이름과 숫자를 따로 떼어서 정리
        for line in f:
            if line.strip() == '':
                continue
            result = line.strip().split(',')
            #name, number를 호출
            self.scores[result[0]] = int(result[1])
        f.close()
    
    #전체 학생의 평균 구하기
    def calculate_average(self):
        total = sum(self.scores.values())
        count = len(self.scores)
        return total / count

    #평균 이상 이름 리스트 반환
    def get_above_average(self):
        #평균 가져오기
        avg = self.calculate_average()
        #만약 점수가 평균 이상이라면 이름을 반환하기
        result = []
        for name, scores in self.scores.items():
            if scores >= avg:
                result.append(name)
        return result

    #평균 미만 학생 저장
    def save_below_average(self):
        #리스트 불러오기
        avg = self.calculate_average()
        #만약 점수가 평균보다 낮다면 평균 미만 학생을 리스트에 저장
        f = open('below_average_korean.txt', 'w')
        for name, score in self.scores.items():
            if avg > score:
                f.write(f'{name},{score}\n')
        f.close()

    #결과 출력
    def print_summary(self):
        avg = self.calculate_average()
        above = self.get_above_average()
        print(f'평균 학생 점수: {avg}')
        print(f'평균 이상 학생: {above}')
```

```python
ss = StudentScores('scores_korean.txt')
ss.print_summary()
ss.save_below_average()
```

```python
#OOP 기반 학생 성적 분석 시스템
class StudentScores():
    #1. 초기 함수
    def __init__(self, filepath):
        self.filepath = filepath
        #점수를 리스트에 저장
        self.scores = {}
        #점수 가져오기
        self.load_scores()

    #2. 파일 불러오기
    def load_scores(self):
        f = open('scores_korean.txt', 'r')
        #리스트의 결과값 저장하기
        for line in f:
            if line.strip() == '':
                continue
            #공백 제거 및 쉼표
            result = line.strip().split(',')
            #리스트에 이름과 점수 넣기
            self.scores[result[0]] = int(result[1])
        f.close()

    #3. 평균 계산하기
    def calculate_average(self):
        #전체 학생 / 학생 수
        total = sum(self.scores.values())
        count = len(self.scores)
        return total / count
    
    #4. 평균 이상 이름 리스트 반환
    def get_above_average(self):
        #평균 가져오기
        avg = self.calculate_average()
        result = []
        #만약 점수보다 높다면 이름 리스트 반환
        for name, scores in self.scores.items():
            if scores >= avg:
                result.append(name)
        return result
            

    #5. 평균 이하 학생 데이터를 별도 파일로 저장하는 메서드
    def save_below_average(self):
        #평균 점수 불러오기
        avg = self.calculate_average()
        #별도 파일명으로 저장
        f = open('below_average_korean.txt', 'w')
        for name, scores in self.scores.items():
            if avg > score:
                f.write(f'{name},{score}\n')
        f.close()

    #6. 결과 출력
    def print_summary(self):
        avg = self.calculate_average()
        above = self.get_above_average()
        print(f'평균 점수는: {avg}점입니다.')
        print(f'평균보다 높은 점수를 받은 학생은 {above}입니다.')

```

```python
ss = StudentScores('scores_korean.txt')
ss.print_summary()
ss.save_below_average()
```
