### 프로그램의 흐름 제어

```python
#if: 분기 ~라면, 조건에 따라 처리 방식을 정의하는 방법
#~라면 => 조건식
#조건식을 잘 만드는것이 중요(매우매우 중요)

#성인 판단 => 성인이면 입장, 아니면(미성년자) 입장 불가
age = int(input('나이를 입력하세요:')) #age > 20

#성인: Say hello, 미성년자: Not allowed.
if age >= 20:
    print("Welcome to the club")
else:
    print("You're not allowed to be here")
```

```python
#짝수 홀수 판단
#짝수와 홀수를 판단 => 조건식 => 짝수인가? 홀수인가?
#num%2==0 -> 짝수, num%2==1 -> 홀수
#참 -> print, 거짓 -> print
number = int(input("정수를 입력하세요: "))

if number % 2 == 0:
    print("짝수입니다.")
else:
    print("홀수입니다.")
```
