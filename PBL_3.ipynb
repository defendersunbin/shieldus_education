# OOP 기반 학생 성적 분석 시스템

```python
#1. StudentScores 클래스 정의
class StudentScores():
    #초기 함수(파일 읽기)
    def __init__(self, filepath):
        self.filepath = filepath
        self.scores = {}
        self.load_scores()

    #파일에서 학생들의 시험 점수 불러오기
    def load_scores(self):
        f = open('scores_korean.txt', 'r')
        #데이터에 있는 이름과 숫자를 따로 떼어서 정리
        for line in f:
            if line.strip() == '':
                continue
            result = line.strip().split(',')
            #name, number를 호출
            self.scores[result[0]] = int(result[1])
        f.close()
    
    #전체 학생의 평균 구하기
    def calculate_average(self):
        total = sum(self.scores.values())
        count = len(self.scores)
        return total / count

    #평균 이상 이름 리스트 반환
    def get_above_average(self):
        #평균 가져오기
        avg = self.calculate_average()
        #만약 점수가 평균 이상이라면 이름을 반환하기
        result = []
        for name, scores in self.scores.items():
            if scores >= avg:
                result.append(name)
        return result

    #평균 미만 학생 저장
    def save_below_average(self):
        #리스트 불러오기
        avg = self.calculate_average()
        #만약 점수가 평균보다 낮다면 평균 미만 학생을 리스트에 저장
        f = open('below_average_korean.txt', 'w')
        for name, score in self.scores.items():
            if avg > score:
                f.write(f'{name},{score}\n')
        f.close()

    #결과 출력
    def print_summary(self):
        avg = self.calculate_average()
        above = self.get_above_average()
        print(f'평균 학생 점수: {avg}')
        print(f'평균 이상 학생: {above}')
```

```python
ss = StudentScores('scores_korean.txt')
ss.print_summary()
ss.save_below_average()
```

```python
#OOP 기반 학생 성적 분석 시스템
#1. class 정의
class StudentScores():
    #초기 함수 정의(파일 위치, 저장할 이름, 성적(점수), 평균 점수 로딩)
    def __init__(self, filepath):
        self.filepath = filepath
        #scores의 경우에는 dictionary로 받아 이름: 점수로 받을 수 있도록
        self.scores = {}
        #점수를 가져오기 위한 초기 함수
        self.generator()

    #생성자: 파일을 읽고 데이터를 딕셔너리에 저장
    def generator(self):
        #먼저 파일을 오픈해야겠지? 그리고 그걸 읽어야겠지? 그 후에는 그 값을 저장해야겠지?
        f = open('scores_korean.txt', 'r')
        result = []
        #데이터셋을 보니 이름,점수로 되어있다. 그러니 일단 공백을 없애고 쉼표로 저장할 수 있도록 해야겠지?
        for line in f:
            result = line.strip().split(',')
            #그 후에는 dictionary에 저장하기 위해서 이름을 [0]번(key)에, 점수를 [1]번(value)에 두도록 하자.
            self.scores[result[0]] = int(result[1])
        f.close()

    #평균 계산하기
    def calculate_average(self):
        #평균을 계산하기 위해서는 먼저 점수를 불러와야한다. Total로 먼저 전체 점수를 불러와야겠지?
        total = sum(self.scores.values())
        #그 후에는 사람들을 불러와야겠지? 그리고 이걸 count해야겠지?
        count = len(self.scores)

        #그 후에는 total / count해야겠지?
        return total / count

    #평균 이상 이름 리스트 반환
    def get_above_average(self):
        #일단 평균에서 계산한 값을 불러와야겠지?
        avg = self.calculate_average()
        result = []
        #만약 평균점수보다 점수가 높다면 이름으로 알려주도록 출력해야겠지?
        for name, scores in self.scores.items():
            if scores >= avg:
                result.append(name)
        return result
    
    #평균 미만 학생 저장
    def save_below_average(self):
        #여기서는 새로운 txt 데이터를 만들어서 평균 점수보다 낮은 학생들을 기록할 수 있도록 해야한다.
        #그러면 먼저 해야할 것은 새로운 데이터를 만들어야한다. 'w'(write)를 사용하자.
        avg = self.calculate_average()
        f = open('below_average_korean.txt', 'w')
        for name, scores in self.scores.items():
            #만약 평균 점수가 점수보다 낮다면 이를 기록해야겠지?
            if avg >= scores:
                #write
                f.write(name)
        f.close()

    #결과 출력
    def print_summary(self):
        #이제 최종 결과를 출력해야한다. 모든 걸 넣으면 된다. 평균, 평균 이상 이름의 리스트를 반환하자.
        avg = self.calculate_average()
        above = self.get_above_average()
        print(f'평균 점수는: {avg}점입니다.')
        print(f'평균 점수보다 높은 점수를 받은 학생은 {above}입니다.')
```

```python
ss = StudentScores('scores_korean.txt')
ss.print_summary()
ss.save_below_average()
```
