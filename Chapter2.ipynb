### 프로그램의 흐름 제어

```python
#if: 분기 ~라면, 조건에 따라 처리 방식을 정의하는 방법
#~라면 => 조건식
#조건식을 잘 만드는것이 중요(매우매우 중요)

#성인 판단 => 성인이면 입장, 아니면(미성년자) 입장 불가
age = int(input('나이를 입력하세요:')) #age > 20

#성인: Say hello, 미성년자: Not allowed.
if age >= 20:
    print("Welcome to the club")
else:
    print("You're not allowed to be here")
```

```python
#짝수 홀수 판단
#짝수와 홀수를 판단 => 조건식 => 짝수인가? 홀수인가?
#num%2==0 -> 짝수, num%2==1 -> 홀수
#참 -> print, 거짓 -> print
number = int(input("정수를 입력하세요: "))

if number % 2 == 0:
    print("짝수입니다.")
else:
    print("홀수입니다.")
```

```python
#if 중첩(elif)
#양수, 음수, 0
num_1 = int(input('숫자를 입력하세요: '))


if num_1 > 0:
     print("양수입니다.")
else:
    #0보다 작거나 같은 숫자
    if num_1 < 0:
        print("음수입니다.")
    else:
        print(0)
```

```python
#elif
num_2 = int(input("숫자를 입력하세요: "))

if num_2 > 0:
    print("양수입니다.")
elif num_2 < 0:
    print("음수입니다.")
else:
    print(0)
```

```python
#운전면허 시험: 70점
score = 69

msg = 'Unknown'

if score >= 70:
    msg = 'Success'
else:
    msg = 'Fail'

print(msg)
```

```python
check_msg = 'Pass'

if score < 70:
    check_msg = 'Fail'
print(check_msg)
```

```python
#반복문: while
#while condition(True/False):
#   반복구문...

#1~5까지의 숫자를 출력
#반복 작업 => 숫자 출력

cnt = 1 #초기값, 카운팅하는 변수
while cnt < 6:
    print(cnt)
    cnt += 1
```

```python
#1~10까지 출력, 홀수만 출력
cnt = 1
while cnt < 11:
    if cnt % 2 == 1:
        print(cnt)
    cnt += 1
```

```python
#continue
#반복문 내부에서 사용 => 이 키워드를 만나면 아래 코드들은 실행되지 않고 블록 종료 => 조건식으로 이동
#홀수를 출력하기

cnt = 0
while cnt < 10:
    cnt += 1
    if cnt % 2 == 0:
        continue
    print(cnt)
```

```python
#break: 반복문 내에서 사용 -> 이 키워드를 만나면 반복문을 종료
#반복문의 조건을 만족했을 때 종료할 수 있도록
cnt = 0
while True:
    #조건식을 두고 break
    cnt += 1
    if cnt > 4:
        break #반복문의 종료
    print(cnt)
```

```python
#for: 시퀸스 데이터를 반복처리할 때 사용
#for val in list_a:
#  변수 val을 이용한 처리

list_a = [1, 2, 3, 4, 5, 100, 200, 300, 1000]
```

```python
#반복문을 통해서 각 요소의 값을 출력
for val in list_a:
    print(val, end=' ')
```

```python
#5명의 학생의 영어 시험 점수를 리스트로 저장
#학생별로 60점 이상이면 pass 출력, 아니면 fail
#총점과 평균 점수 출력

score = [80, 50, 70, 100, 55]
snum = 0
sum = 0

for s in score:
    snum += 1
    check = 'Pass' if s >= 60 else 'Fail'
    print(snum, s, check)
    sum += s

print('-'*30)
print('총점: ', sum)
print('평균: ', sum//len(score))
```

```python
#range 함수: 일련번호 담은 리스트를 생성 -> 반환
#range(start, end, step)

for num in range(11): #(0, 10, 1)
    print(num, end=' ')
```

```python
for num in range(1, 11, 2):
    print(num)
```

```python
#구구단

for num in range(1, 10):
    print(f'2 X {num} = {2*num}')


for num in range(1, 10):
    print(f'3 X {num} = {3*num}')

for num in range(1, 10):
    print(f'4 X {num} = {4*num}')
```

```python
#단을 표현하는 숫자
for dan in range(2, 5):
    print(dan, '단')
    for num in range(1, 10):
        print(f'{dan} X {num} = {dan * num}') #2 X {num} 에서 2를 {dan}으로 해줘야 다음 단으로 구구단을 출력할 수 있다.
```
