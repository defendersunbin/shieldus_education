# Class

### Class의 상속

```python
#사칙연산을 하는 클래스
# => 클래스 이름, 변수, 함수(메서드), 초기화 함수(생성자)
#클래스의 이름 대문자로 시작 => FourCal 
# 캐멀케이스(CamalCase)
#변수: 2개
#함수(메서드): 설정: +, *, -, //, %
class Calculator():
    #클래스의 멤버: 변수, 메서드
    #메서드를 통해 변수(인스턴스 변수)를 생성 => 클래스를 통해서 객체로 만들면 변수가 메모리에 로드

    #초기화 함수 -> 변수 초기화
    def __init__(self, first, second):
        self.first = first
        self.second = second

    #함수 생성시에 self는 강제(없으면 메모리 참조가 불가능함)
    #self 사용이유: 인스턴스 멤버를 사용하기 위해서
    def setData(self, first, second): #self: Instance 자체
        #인스턴스 생성과 데이터 할당
        self.first = first #메서드의 치역 변수(self.first와 first는 다르다.)
        self.second = second

    def printData(self):
        print(f'first: {self.first}, second: {self.second}')

    def add(self):
        return self.first + self.second

    def sub(self):
        return self.first - self.second
    
    def mul(self):
        return self.first * self.second
    
    def div(self):
        return self.first / self.second
```

```python
#클래스의 상속: 새로운 클래스를 만든다.
#상속 => 기존 클래스의 멤버들을 받아서 내것으로 사용
#오버라이딩 => 메서드를 수정해서 사용, 처리 블록을 수정해서 사용
#MoreFourCal: 새로운 클래스 정의 => pow 제곱을 구해주는 함수를 추가, div 메서드 수정
class MoreFourCalculator(Calculator):
    # pass
    def pow(self):
        return self.first**self.second
    
    #오버라이딩: 함수의 이름과 매개변수는 그대로 유지하고 => 처리부분만 수정해서 정의
    def div(self):
        # return super().div() #Calculator div 메서드 호출
        if self.second == 0:
            return 0
        return self.first / self.second
```

```python
#cal1 = MoreFourCalculator : 에러 => 생성자 호출
cal1 = MoreFourCalculator(3, 0)
```

```python
type(cal1)
```

```python
cal1.printData()
```

```python
cal1.pow()
```

```python
cal1.div()
```
