# HTML Parsing

```python
from bs4 import BeautifulSoup

html_doc = """
<html>
    <head>
        <title>Example Page</title></head>
    <body>
        <h1>Welcome to Web Scraping</h1>
        <p class="description">This is an example paragraph.</p>
        <a href="https://example.com">Visit Example</a>
    </body>
</html>
"""
#Beautiful Soup 객체
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.title.string)
print(soup.h1.string)
print(soup.a['href'])
print(soup.p['class'])
```

```python
#find (tag), find_all (tag)
a_tag = soup.find('a')
print(a_tag)
a_tags = soup.find_all('a')
print(a_tags)
```

```python
links = []
for a_tag in a_tags:
    links.append(a_tag['href'])

print(links)
```

```python
#select (selector) ==> list
#select_one(selector) => tag
#class="description title big small"
#.description {} .title {} .big {} .small {}
#id="first" => #id{}
#tag 선택
tag1 = soup.select_one('.description')
print(tag1)
tag2 = soup.select_one('.description')
print(tag2)
```

# SQLite (관계형 데이터베이스 시스템 RDBMS)

```python
import sqlite3

#데이터 연결
conn = sqlite3.connect('first.db')
print('데이터베이스 연결 성공')
conn.close
```

```python
#관계형 데이터베이스의 특징 => 테이블 형식으로 데ㅣㅇ터 관리
#데이터를 저장할 공간을 생성 => 테이블 생성
sql_create = '''
CREATE TABLE IF NOT EXISTS users (
id INTEGER PRIMARY KEY AUTOINCREMENT,
name TEXT NOT NULL,
age INTEGER,
city TEXT
)
'''
```

```python
#1. 데이터베이스 연결 => Connection
conn2 = sqlite3.connect('first.db')
print('데이터베이스가 연결되었습니다.')
```

```python
#2. SQL 실행 객체 => cursor(): SQL 실행결과를 메모리에 저장하는 것
cursor = conn2.cursor()
```

```python
#3. SQL 실행: execute(), executemany()
cursor.execute(sql_create)
```

```python
#4. close()
conn2.close()
print('데이터베이스가 종료되었습니다.')
```

```python
#데이터 입력 -> 행 단위 입력
#id, name, age, city
#sql
sql_insert = "INSERT INTO users(name, age, city) VALUES(?, ?, ?)"

#데이터베이스 연결
conn_1 = sqlite3.connect('first.db')
print('데이터베이스가 연결되었습니다.')

#cursor로 실행결과 메모리에 저장
cursor = conn_1.cursor()

#SQL 실행
cursor.execute(sql_insert, ('Lionel Messi', 39, 'Buenos Aires'))

#변경사항 저장: 입력, 수정, 삭제 => 완료 또는 초기화
#Commit
conn_1.commit()
if conn_1.commit():
    print('데이터가 성공적으로 저장했습니다.')
else:
    print('데이터 저장에 실패했습니다.')

#sql 실행종료
conn_1.close()
print('데이터베이스가 종료되었습니다.')
```

```python
#SQL 행단위 검색: 질의
sql_select = 'SELECT * FROM users'
sql_select2 = 'SELECT name, city FROM users WHERE age > ?'

#SQL 실행
conn_2 = sqlite3.connect('first.db')
cursor = conn_2.cursor()
cursor.execute(sql_select2, (39, ))

#select 결과 => 새로운 table
#fetchall() => 전체 행 반환
#fetchone() => 하나의 행 반환
rows = cursor.fetchall()
# print(rows)

for row in rows:
    print(row)

conn_2.close()
print('데이터베이스가 종료되었습니다.')
```

```python
#sql
sql_insert = "INSERT INTO users(name, age, city) VALUES(?, ?, ?)"

#데이터베이스 연결
conn_3 = sqlite3.connect('first.db')

#데이터
data = [
    ('Henry', 50, 'France'),
    ('Zlatan', 45, 'Sweden'),
    ('Neymar', 34, 'Brazil')
]

#cursor
cursor = conn_3.cursor()

#sql실행
cursor.executemany(sql_insert, data)

#Commit
conn_3.commit()
print('데이터베이스에 성공적으로 저장했습니다.')

#sql 실행종료
conn_3.close()
print('데이터베이스가 종료되었습니다.')
```

```python
#SQL 행단위 검색: 질의
sql_select = 'SELECT * FROM users'

#SQL 실행
conn_2 = sqlite3.connect('first.db')
cursor = conn_2.cursor()
cursor.execute(sql_select)

#select 결과 => 새로운 table
#fetchall() => 전체 행 반환
#fetchone() => 하나의 행 반환
rows = cursor.fetchall()
# print(rows)

for row in rows:
    print(row)

conn_2.close()
print('데이터베이스가 종료되었습니다.')
```

```python
#데이터 수정
#행을 찾고 => 특정 컬럼의 데이터를 수정
sql_update = "UPDATE users SET age=?, name=? WHERE id=?"

#데이터베이스 연결
conn_3 = sqlite3.connect('first.db')

#cursor
cursor = conn_3.cursor()
cursor.execute(sql_update, (45, 'Ji Sung', 5))

conn_3.commit()
conn_3.close()
```

```python
#데이터 삭제
#행을 찾고 => 특정 컬럼의 데이터를 삭제
sql_delete = "DELETE FROM users WHERE id=?"

conn_4 = sqlite3.connect('first.db')
cursor = conn_4.cursor()
cursor.execute(sql_delete, (7, ))
conn_4.commit()
conn_4.close()
```
