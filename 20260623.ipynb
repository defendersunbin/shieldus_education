```python
#exam - 1
#과목별 점수, 총점과 평균

#변수: 데이터를 가리킨다. => 프로그램 내부에서 필요할 때 다시 사용할 수 있도록 하기 위해서
k_score = int(input('국어 점수를 입력하세요: '))
e_score = int(input('영어 점수를 입력하세요: '))
m_score = int(input('수학 점수를 입력하세요: '))
```

```python
#총점과 평균
total_score = k_score + e_score + m_score #숫자 타입의 연산을 이용해서 합을 구함
avg_score = total_score // 3
```

```python
#결과 출력
#print(f"총점: {total_score}, 평균: {avg_score}")
print('성적표')
print('*' * 30)
print(f"국어:", k_score, '점')
print('영어: ' + str(e_score) + '점')
print(f"수학: {m_score}점")
print('-' * 30)
print(f'총점: {total_score}점')
print(f'평균: {avg_score}점')
```

```python
#18살 이상 그리고 남성 판별하는 조건식
#age >= 18, gender = 'male', 그리고 => and
#변수
gender = 'M' # 'F'
age = 20

#비교연산
check = gender == 'M' and age >= 18

#결과 출력
print('성인 남성 판별: ', check)
print(f'성인 남성 판별: {check}')
```
