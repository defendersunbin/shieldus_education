# 데이터 전처리 EX-5

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder

data = pd.read_csv('winequality-red.csv')

data
```

```python
#표준화
standard_scaler = StandardScaler()
category_cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides']
data[category_cols] = standard_scaler.fit_transform(data[category_cols])

#시각화 자료 출력
#데이터 확인
plt.figure(figsize=(10, 5))
#제목
plt.title('Wine Quality')
#x축
plt.xlabel('Index')
#Y축
plt.ylabel('Value')
#data1
plt.plot(data['fixed acidity'], label='Fixed Acidity', marker='o')
#data2
plt.plot(data['volatile acidity'], label='Volatile Acidity', marker='o')
#data3
plt.plot(data['citric acid'], label='Citric Acidity', marker='o')
#data4
plt.plot(data['residual sugar'], label='Residual Sugar', marker='o')
#data5
plt.plot(data['chlorides'], label='Chlorides', marker='o')
plt.legend()
plt.show()
```
