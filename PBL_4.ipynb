```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#1. 먼저 SalesAnalysis 클래스를 정의해야지
class SalesAnalysis:
    #초기 함수를 정의하자
    def __init__(self):
        self.dates = pd.date_range(start='2024-01-01', end='2024-12-31', freq='ME')
        self.sales = np.random.randint(low=1000, high=10000, size=len(self.dates))

    #날짜 생성하기
    def monthly_summary(self):
        #2024년 1월~12월까지의 날짜 생성 - Dictionary 형태
        df = pd.DataFrame({
            'date': self.dates,
            'sales': self.sales
        })

        #월에 대한 컬럼 추가
        df['month'] = df['date'].dt.month
        #월별 합계
        self.monthly = df.groupby('month')['sales'].sum()
        return self.monthly

    #시각화된 자료로 만들기
    def visualize(self):
        #월별 데이터 준비
        self.monthly_summary()
        #크기 조절
        plt.figure(figsize=(10,5))
        #꺾은선 그래프
        plt.plot(self.monthly.index, self.monthly.values, marker='o')
        #제목
        plt.title('Business Insight')
        #x축 라벨
        plt.xlabel('Month')
        #y축 라벨
        plt.ylabel('Sales')
        #시각화 최종 자료 출력
        plt.show()
```

```python
sa = SalesAnalysis()
sa.visualize()
```
