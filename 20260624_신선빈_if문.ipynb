# if문

### 숫자를 입력받아 짝수, 홀수를 판별하는 프로그램

```python
#짝수 홀수 판단
#짝수와 홀수를 판단 => 조건식 => 짝수인가? 홀수인가?
#num%2==0 -> 짝수, num%2==1 -> 홀수
#참 -> print, 거짓 -> print
num1 = int(input("정수를 입력하세요: "))

if num1 % 2 == 0:
    print("짝수입니다.")
else:
    print("홀수입니다.")
```

### 숫자를 입력받아 양수, 음수, 0을 판별하는 프로그램

```python
num2 = int(input("숫자를 입력하세요: "))

if num2 > 0:
    print("양수")
elif num2 < 0:
    print("음수")
else:
    print(0)
```

### 독감예방 접종이 가능한지 여부를 확인하는 프로그램

```python
# 15세 미만 혹은 65세 이상의 경우 무료 예방접종 가능 메시지 출력


age = int(input("나이(연세)를 입력하세요: "))
current_year = 2026
born = current_year - age

#1. 15세 미만은 접종이 불가능, 그 이후 나이부터 65세 이상의 경우 무료 예방접종 가능
if 15 < age or age >= 65:
    print("무료 예방접종 대상자입니다.")
else:
    print("무료 예방접종 대상자가 아닙니다.")

#2. 성인은 2년마다 건강검진을 받을 수 있음.
if age >= 20:
    print("성인은 2년마다 건강검진을 받을 수 있습니다.")

#3. 짝수 해에 태어난 사람은 올해가 짝수년이라면 검사 대상이 된다.
if born % 2 == 0:
    print("검사 대상입니다.")

#4. 40세 이상의 경우, 암 검사도 무료로 검사할 수 있음.
if age >= 40:
    print("무료 암 검사를 진행할 수 있습니다.")
```
