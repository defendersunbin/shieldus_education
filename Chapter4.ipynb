# 기초통계

```python
#pandas를 이용 => 평균, 중앙, 최빈값
#mean, median, mode
import pandas as pd
```

```python
data = [10, 15, 10, 20, 30, 10, 40]

data_series = pd.Series(data)
```

```python
#평균
mean_val = data_series.mean()
#중앙값
median_val = data_series.median()
#최빈값(만약 그 중 첫번째 값을 알고 싶다면 mode()[0]을 붙이면 된다)
mode_val = data_series.mode()[0]
```

```python
print(mean_val)
print(median_val)
print(mode_val)
```

# 데이터 분산성(분산, 표준편차, 사분위수)

```python
data = {'점수': [55, 60, 65, 70, 75, 80, 85, 90, 95]}

df = pd.DataFrame(data)
```

```python
#분산 - var
var_value = df['점수'].var()
print(var_value)
```

```python
#표준편차 - std
std_value = df['점수'].std()
print(std_value)
```

```python
#사분위수
q1_value_pd = df['점수'].quantile(0.25) #1사분위수
q2_value_pd = df['점수'].quantile(0.5) #2사분위수
q3_value_pd = df['점수'].quantile(0.75) #3사분위수
q4_value_pd = df['점수'].max() #최대값

print(q1_value_pd)
print(q2_value_pd)
print(q3_value_pd)
print(q4_value_pd)

#IQR (Q3-Q1)
iqr_value = q3_value_pd - q1_value_pd
print(iqr_value)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statistics
import numpy as np
import platform

# 한글 폰트 설정
# Windows
# plt.rcParams['font.family'] = 'Malgun Gothic'

# Mac 한글 폰트
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

#Seaborn 기본 스타일 설정
sns.set_theme(style="whitegrid", font="AppleGothic" if platform.system()=='Darwin' else "Malgun Gothic")
print("한글 시각화 준비 완료")
```

```python
#데이터 준비
x = [1, 2, 3, 4, 5] #x축 데이터
y = [2, 3, 5, 7, 11] #y축 데이터

#그래프 생성
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Prime Numbers')

#그래프 꾸미기
#그래프 제목
plt.title("Line Graph Example")
#x축
# plt.xlabel('')
```

```python
#bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]

#시각화
plt.bar(categories, values)
# plt.barh(categories, values)
```

```python
#히스토그램(histogram)
data = np.random.randn(1000)

#히스토그램 차트
plt.hist(data, bins=10, edgecolor='black')
```

# Seaborn

```python
#학생 점수 데이터
scores = [56, 67, 45, 89, 90, 92, 76, 58, 70, 73, 85]

#Seaborn 히스토그램 생성
sns.histplot(scores, bins=5, kde=True, color='blue', edgecolor='black') #kde=True는 밀도값, bins=5 5개 구간으로 구분

plt.title('학생 점수 분포')
plt.xlabel('점수 구간')
plt.ylabel('학생 수')
plt.show()
```
